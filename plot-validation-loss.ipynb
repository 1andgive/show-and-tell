```python
from __future__ import division
import matplotlib.pyplot as plt
import utils
import numpy as np
%matplotlib inline
```

```python
def plotter(pkl_file_1,pkl_file_2,output_filename):
    loss_dict_1 = utils.unpickle(pkl_file_1)
    loss_dict_2 = utils.unpickle(pkl_file_2)
    losses_val_1 = loss_dict_1['losses_val']
    losses_val_2 = loss_dict_2['losses_val']

    fig = plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    
    # plotting training and val together requires shortening losses_train to 76 * 124
    y1 = np.array(losses_val_1) / 128
    y2 = np.array(losses_val_2) / 128
    plt.plot(y1,'r')
    plt.plot(y2,'b')

    plt.gca().set_ylim([1,6])

    # LSTM memory size
    plt.title('Loss, regular RNN')
    plt.legend(('state_size=256', 'state_size=512'),loc='upper right')
    
    fig.savefig(output_filename)

plotter('2017-12-27_checkpoints_experiment_elman_256/losses.pkl',
        '2017-12-24_checkpoints_experiment_elman_512/losses.pkl',
        'output.png')
```
